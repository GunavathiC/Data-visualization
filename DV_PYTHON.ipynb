Pie Chart

```python
#4. Pie chart(Category Distribution)
category_counts=df["Category"].value_counts()
plt.pie(category_counts,labels=category_counts.index,autopct="%1.0f%%")
plt.title("Category Distribution (Matplotlib)")
plt.show()
```

**Histogram**

```python
#5. Histogram (Growth Distribution)
plt.hist(df["Growth"], bins=5, color="red",edgecolor = "black")
plt.title("Growth Distribution(Matplotlib)")
plt.xlabel("Growth")
plt.ylabel("Frequency")
plt.show()
```

**Seaborn Visualization**

```python
sns.lineplot(x="Year",y="Sales",data=df,marker="o",color="red")
plt.title("Sales over Years (Seaborn)")
plt.show()
```

```python
sns.barplot(x="Category",y="Profit",data=df,color="red")
plt.title("Sales over Years (Seaborn)")
plt.show()
```

```python
sns.regplot(x="Sales", y ="Profit",data=df,color="red")
plt.title("Sales vs Profit(Seaborn)")
plt.show()
```
